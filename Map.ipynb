```python
import pandas as pd
import folium
```

```python
df=pd.read_excel(open('data/VA Healthcare Facilities.xlsx', "rb"))
df2=pd.read_excel(open('data/VA Surgical Facilities.xlsx', "rb"))
df3=pd.read_excel(open('data/DoD Surgical Facilities.xlsx', "rb"))
df4=pd.read_excel(open('data/Joint Venture Sites.xlsx', "rb"))
state_data=pd.read_excel(open('data/vet_pop_state.xlsx', "rb"))

# Define Various Global Variables

location=[47.1164,-101.2996]
locationlist=df[["Latitude","Longitude"]].values.tolist()
surgicallist=df2[["Latitude","Longitude"]].values.tolist()
dodlist=df3[["Latitude","Longitude"]].values.tolist()
jvlist=df4[["Latitude","Longitude"]].values.tolist()
labels=df["Name"].values.tolist()
surglabels=df2["Name"].values.tolist()
dodlabels=df3["Name"].values.tolist()
jvlabels=df4["Name"].values.tolist()

map=folium.Map(location=location,zoom_start=3,tiles="CartoDB positron")
folium.TileLayer('openstreetmap').add_to(map)
folium.TileLayer('stamenterrain').add_to(map)

map.choropleth(
        geo_data='data/us_states.json',
        name='Veteran Population by State',
        data=state_data,
        columns=['State','Veteran Population'],
        key_on='feature.id',
        fill_color="YlGn",
        fill_opacity=0.7,
        line_opacity=0.2,
        legend_name="Veteran Population by State"
)

from folium.plugins import MeasureControl
map.add_child(MeasureControl())

fg=folium.FeatureGroup(name="All VA Healthcare Facilities", show=False)

for point in range(len(locationlist)):
    popup=folium.Popup(labels[point],parse_html=True)
    fg.add_child(folium.RegularPolygonMarker(
        locationlist[point],
        popup=popup,
        fill_color='#EE1C25',
        number_of_sides=5,
        radius=10,
        )).add_to(map)

map.add_child(fg)

fg2=folium.FeatureGroup(name="VA Surgical Facilities", show=False)

for point in range(len(surgicallist)):
    popup=folium.Popup(surglabels[point],parse_html=True)
    fg2.add_child(folium.RegularPolygonMarker(
        surgicallist[point],
        popup=popup,
        fill_color='blue',
        number_of_sides=5,
        radius=10,
        )).add_to(map)

map.add_child(fg2)

fg3=folium.FeatureGroup(name="DoD Surgical Facilities", show=False)

for point in range(len(dodlist)):
    popup=folium.Popup(dodlabels[point],parse_html=True)
    fg3.add_child(folium.RegularPolygonMarker(
        dodlist[point],
        popup=popup,
        fill_color='green',
        number_of_sides=3,
        radius=15,
        )).add_to(map)

map.add_child(fg3)

fg4=folium.FeatureGroup(name="DoD/VA Joint Venture Sites", show=False)

for point in range(len(jvlist)):
    popup=folium.Popup(jvlabels[point],parse_html=True)
    fg4.add_child(folium.RegularPolygonMarker(
        jvlist[point],
        popup=popup,
        fill_color='yellow',
        number_of_sides=6,
        radius=10,
        )).add_to(map)

map.add_child(fg4)

fg5=folium.FeatureGroup(name="DoD/VA Shared Sites", show=False)

map.add_child(fg5)

map.add_child(folium.LayerControl())
```
